```python
import pandas as pd
import numpy as np
```

```python
# 영화 정보 (id)
movies = pd.read_csv("http://grepp-reco-test.s3.ap-northeast-2.amazonaws.com/tmdb_5000_movies.csv")
```

```python
#연출/배우 정보 (movie id) 위의 id와 movie id를 조인 할거임
credits = pd.read_csv("http://grepp-reco-test.s3.ap-northeast-2.amazonaws.com/tmdb_5000_credits.csv")
```

```python
movies.head()

#인기도를 따질 때 popularity와 vote 가 가능하지만 일단 popularity로 할거임
```

```python
movies.shape
#4803개의 영화와 20개의 정보가 들어가 있다
```

```python
import json

#장르네임 칼럼을 새로 만들거고 [{"id": 12, "name": "Adventure"}, {"id": 14, ".. 이렇게 표시되어있는 형태에서
# adventure 같은 name 필드에 있는 값만 가져올것이다

def add_genre_name(j):
    genres=[]
    ar = json.loads(j)
    for a in ar:
        genres.append(a.get("name"))
    return " ".join(sorted(genres))

#위에서 만든 장르이름만 추출한 것을 기존 movies에 장르 열을 만들어 붙인다
movies[' genres_name '] = movies.apply(lambda x: add_genre_name(x.genres), axis=1) 
```

```python
credits.head() #여기 movie id와 위의 movie의 id를 조인하면 영화에 대한 완전한 정보를 얻을 수 있음
```

```python
credits.shape
```

movies와 credits 데이터 프레임을 판다스의 merge를 이용해 조인하자

```python
movie_credits= pd.merge(movies, credits, left_on ='id', right_on ='movie_id')
```

```python
#title_x와 같이 _x가 되어있는건 칼럼명이 중복되어서 그런거임
#필요 없는 칼럼을 drop을 사용해 날려보자


movie_credits = movie_credits.drop(columns =['homepage', 'title_x', 'title_y', 'status', 'production_countries', 'production_companies' ])
```

```python
movie_credits.info() #값이 비어있는 코드가 있는지 확인 -> overview 3개, tagline 400개 정도 없음....
```

```python
movie_credits.describe() #describe로 수치값들의 통계를 구함
```

```python
#popularity를 기준으로 추천을 할거니까 ascending을 false로 해서 내림차순을 만들음

popularity = movie_credits.sort_values('popularity',ascending=False)
```

```python
#시각화에 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
plt.figure(figsize=(12,6))
ax= sns.barplot(
    x=popularity[ 'popularity' ].head(10),#인기
    y=popularity[ 'original_title' ].head(10) #영화정보
)
plt.title('most popular by popularity', weight='bold')
plt.xlabel( 'score of popularity', weight = 'bold')
plt.savefig('best_popular_movies.png')
```

```python
#장르 정보가 들어오면 그 장르 안에서 인기도가 높은것들만 리턴하고 없음 걍 original title을 popularity 순으로 헤드 리턴
#입력 장르가 들어왔다면 contains를 써서 만들어 뒀던 장르네임 컬럼에 부분매칭인 substring으로 있는지 즉 존재가 하는지 확인해 있으면(조건을 만족하면) 그걸 헤드리턴
#하는 함수를 만들었다

def reco_top_scored_one(n, genre=None):
    if genre is None:
        return popularity["original_title"].head(n)
    else:
        return popularity[popularity[' genres_name '].str.contains(genre)]["original_title"].head(n)
```

```python
print(reco_top_scored_one(10))
```

```python
#science fiction이라는 장르의 탑텐을 선정하고 싶다 -> 장르값 입력 됨

print(reco_top_scored_one(10, "Sceince Fiction"))
```
