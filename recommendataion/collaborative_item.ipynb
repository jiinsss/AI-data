```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#우선 더미 데이터들을 가지고 해보자
dummy_rating = pd.read_csv("http://grepp-reco-test.s3.ap-northeast-2.amazonaws.com/dummy_rating.csv", index_col =0)
```

```python
dummy_rating.head()
```

```python
#아이템은 행이 유저가 아니라 아이템이어야하니가 행과 열을 치환시켜준다

dummy_rating = dummy_rating.T
```

```python
dummy_rating.head()
```

```python
#결측치는 나중에 에러 만드니까 숫자 0으로 대체
dummy_rating.fillna(0, inplace =True)
dummy_rating
```

```python
#평점 정보를 보정, 이후에 코사인 유사도를 사용하면 이는 피어슨 유사도였다
#평균으로부터의 차이
def standardize(row):
    new_row = (row -row.mean())/(row.max()- row.min())
    return new_row
#평균보다 높으면 positive 낮으면 negative

dummy_rating_std = dummy_rating.apply(standardize)
dummy_rating_std.head()
#user1 은 sifi 좋 come 싫
```

```python
#정규화 기반 아이템간의 유사도 측정
corrMatrix= pd.DataFrame(cosine_similarity(dummy_rating_std), index= dummy_rating.index, columns=dummy_rating.index)
corrMatrix

#scifi 1은 sicif 2, 3는 양수가 comedy1,2,3와는 음수가 나왔다
```

```python
def get_similar(movie_name) :
    similar_score =corrMatrix[movie_name]
    #앞서 보정된 값을 가지고 평점의 내림차순으로 정렬
    simliar_score= similar_score.sort_values(ascending =False)
    return similar_score
```

```python
#scifi_lover = "user1"
movie_i_liked = "scifi1"

similar_scores = pd.DataFrame(get_similar(movie_i_liked))
similar_scores.head(10)
```
